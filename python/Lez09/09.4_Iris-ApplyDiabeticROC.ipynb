# Logistic regression

```python
from sklearn.linear_model import LogisticRegression # algoritmo lineare di classificazione
from sklearn.model_selection import train_test_split
```

```python
from IPython.display import Image
Image(url='https://frenzy86.s3.eu-west-2.amazonaws.com/fav/iris_8.png',width=900,height=300)
```

```python
#Split the data into 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=667
                                                    )
print(y_test)
```

```python
#Train the model with Logistic Regression
model = LogisticRegression()
model.fit(X_train, y_train) #Training the model
```

```python
model.coef_ # come mai 12 coeff
```

```python
## How classification work
## X_test.iloc[0]
X1 = 6.5 # sepal length
X2 = 3.0 # sepal width
X3 = 5.5 # petal length
X4 = 1.8 # petal width

# presi da sopra 
x1=(10.36122821-0.47684664*X1 + 0.76508588*X2 -2.3968327*X3 -0.96721658*X4) # setosa
x2=(2.10475101+0.41151869*X1 -0.18774797*X2 -0.16175093*X3 -0.84080686*X4) # versicolor
x3=(-12.46597922+0.06532795*X1 -0.57733791*X2 +2.55858363*X3 +1.80802344*X4) #  verginica

p1 = 1/(1+np.exp(-x1)) # prob setosa
p2 = 1/(1+np.exp(-x2)) # prb versicolor
p3 = 1/(1+np.exp(-x3)) #prob virginica

prob = np.array([p1,p2,p3]) # non sommano ad 1, abbiamo uno spazio di probabilitÃ  senza mutua esclusione degli eventi


# abbiamo il vincolo di mutua esclusivitÃ , lo aggiunguamo con softmax
def softmax(x):
    """Applying softmax to the array x """
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum(axis=0)

prob_normalized = softmax(prob)
prob_normalized

# per verificare la somma ad uno usiano np.sum(prob_normalized)
```

Where p1â€‹, p2â€‹, and p3â€‹ are the probabilities for each class calculated using the logistic function.

The probabilities are calculated using the sigmoid function (the logistic function) to map the output x values to a range between 0 and 1:

$$p = \frac{1}{1 + e^{-x}}$$
However, the probabilities you've computed might not sum up to 1 due to how you've calculated and normalized them. The probabilities of each class for a given observation should sum up to 1 after the application of the softmax function.

To ensure the probabilities are correctly normalized, you should apply the softmax function to the output scores x for each class:

$$\text{softmax}(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{N} e^{x_j}}$$
where N is the total number of classes.

The softmax function will ensure that the probabilities sum up to 1 across all classes for a given observation. This normalization is often used in multi-class classification problems

```python
from sklearn.linear_model import LogisticRegression # il mio problema Ã¨ di classificazione, il linear regression Ã¨ un errore

model_lr= LogisticRegression()
model_lr.fit(X_train,y_train)
y_pred = model_lr.predict(X_test) # crea le predizioni
y_pred

from sklearn.metrics import confusion_matrix


labels = ['Iris-virginica','Iris-setosa', 'Iris-versicolor']

cm = confusion_matrix(y_test, y_pred) # y_true==y_pred
sns.heatmap(cm , square=True, annot=True, cbar=False,cmap='Blues',xticklabels=labels,yticklabels=labels)
plt.title('Confusion Matrix')
plt.xlabel('Actual')
plt.ylabel('Prediction');

from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier # altro classificatore, attenzione al seed

model_rfc = RandomForestClassifier(n_estimators=100, max_depth=5)
model_rfc.fit(X_train,y_train)
y_pred = model_rfc.predict(X_test) # crea le predizioni
y_pred

from sklearn.metrics import confusion_matrix


labels = ['Iris-virginica','Iris-setosa', 'Iris-versicolor']

cm = confusion_matrix(y_test, y_pred) # y_true==y_pred
sns.heatmap(cm , square=True, annot=True, cbar=False,cmap='Blues',xticklabels=labels,yticklabels=labels)
plt.title('Confusion Matrix')
plt.xlabel('Actual')
plt.ylabel('Prediction');

from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

# ROC

```python
from sklearn.metrics import RocCurveDisplay # implementato nell'anno scorso

plt.figure(figsize=(8,8))
ax = plt.gca()

RocCurveDisplay.from_estimator(model_lr, X_test, y_test,ax=ax) # regression
RocCurveDisplay.from_estimator(model_rfc, X_test, y_test,ax=ax) # random forest
plt.plot([0,1],[0,1],'--b')

plt.title("Receiver Operating Characteristic (ROC) curves")
plt.legend(loc=4);
```
