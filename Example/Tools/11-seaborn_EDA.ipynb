A data frame with 244 observations on the following 8 variables.


- **total_bill:** a numeric vector, the bill amount (dollars)
- **tip:** a numeric vector, the tip amount (dollars)
- **sex:** factor with levels Female Male, gender of the payer of the bill
- **smoker:** factor with levels No Yes, whether the party included smokers
- **day:** factor with levels Friday Saturday Sunday Thursday, day of the week
- **time:** factor with levels Day Night, rough time of day
- **size:** numeric vector, number of people in party



# TIPS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
preview_window=2

tips = pd.read_csv('https://frenzy86.s3.eu-west-2.amazonaws.com/fav/tips.csv')
print("Preview ------------------------")
print(tips[:preview_window])

print("")
print("Descrittiva")
print(tips.describe().T)


print("")
print("Info")
print(tips.info())


print("")
print("Null find")
print(tips.isnull().sum()) # ritorna dataframe Boolean, True è NaN, li somma


print("\n\n")
print("smoker count ------------------------")
#tips['smoker'].value_count()
#tips['smoker'].value_count(normalize=True) # percentuale
sns.countplot(x='smoker',data=tips,hue='smoker').set(title="smoker count");
plt.show()

print("\n\n")
print("sex count ------------------------")
sns.countplot(x='sex',data=tips,hue='sex').set(title="sex count");
plt.show()

print("\n\n")
print("day count ------------------------")
sns.countplot(x='day',data=tips,hue='day').set(title="day count");
plt.show()


print("\n\n")
print("Launch dinner count ------------------------")
plt.figure(figsize=(6,3))
sns.countplot(x='time',data=tips,hue='time').set(title="Launch dinner count");
plt.show()


print("\n\n")
print("Launch dinner count trasposed ------------------------")
plt.figure(figsize=(6,3))
sns.countplot(y='time',data=tips,hue='time').set(title="Launch dinner count trasposed");
plt.show()

print("\n\n")
print("Eda ------------------------") # si può fare in base alle classi come visto sopra
sns.pairplot(tips,height=4,corner=True);
plt.show()


print("\n\n")
print("Eda class colorcode Categorical = 'sex', 'smoker' , 'day', 'time' , 'size' ------------------------")
sns.pairplot(tips,height=3,hue='sex');
plt.show()


print("\n\n")
print("Eda class colorcode change hue smoker ------------------------")
sns.pairplot(tips,height=3,hue='smoker');
plt.show()


print("\n\n")
print("Eda class colorcode change hue day ------------------------")
sns.pairplot(tips,height=3,hue='day');
plt.show()


print("\n\n")
print("Eda class colorcode change hue time ------------------------")
sns.pairplot(tips,height=3,hue='time');
plt.show()


print("\n\n")
print("Eda class colorcode change hue size ------------------------")
sns.pairplot(tips,height=4,hue='size');
plt.show()


print("\n\n")
print("time,sex,bill,tip ------------------------")
sns.relplot(x="total_bill", y="tip", hue="day",  col="time", row="sex", data=tips);
plt.show()


print("\n\n")
print("All Variables ------------------------")
sns.relplot(x="total_bill", y="tip", hue="day", col="time", row="sex", size="size",  style="smoker", data=tips);
plt.show()


print("")
print("bill tip time")
sns.relplot(x="total_bill", y="tip", col="time",hue="smoker", size="size", data=tips);
plt.show()

print("")
print("bill tip " )
sns.relplot(x="total_bill", y="tip", hue="time",col="day", data=tips) ;
plt.show()

print("")
print("bill tip wrap " )
sns.relplot(x="total_bill", y="tip", hue="time",col="day", col_wrap=2, data=tips) ;
plt.show()

print("")
print("day bill height catplot" )
sns.catplot(x="day", y="total_bill", hue="smoker",kind="swarm", data=tips, height=8, aspect=1);
plt.show()

print("")
print("day bill violin" )
sns.catplot(x="day", y="total_bill", hue="smoker",kind="violin", split=True, data=tips, height=8,aspect=1);
plt.show()


print("")
print("bill scatterplot" )
plt.figure(figsize=(19,8))
sns.scatterplot(x="total_bill", y="tip", data=tips);
plt.show()


print("")
print("tip histplot" )
plt.figure(figsize=(19,8))
#sns.distplot(tips['tip']); # deprecated
sns.histplot(tips['tip'], kde=True);
plt.show()


print("")
print("day barplot" )
plt.figure(figsize=(19,8))
sns.barplot(x="day", y="tip", data = tips);
plt.show()

print("")
print("boxplot day tip scatterplot bill day" )
fig,ax =plt.subplots(1, 2, sharey=True, figsize=(16, 10))
sns.boxplot(x="day", y="tip", data=tips, ax=ax[0])
sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips, ax=ax[1]);
plt.show()


print("")
print("regplot bill tip" )
plt.figure(figsize=(19,8))
sns.regplot(x="total_bill", y="tip", data=tips);
plt.show()


print("")
print("jointplot bill tip" )
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg',height=16)
plt.show()

print("")
print("corr " )
corr = tips.corr(numeric_only=True)
print(corr[:preview_window])


print("")
print("subplots heatmap")
plt.subplots(figsize=(8,6))
sns.heatmap(corr, annot = True);
plt.show()


print("")
print("pairplot reg")
#sns.pairplot(data=tips,kind="reg", size=4); # deprecated
sns.pairplot(data=tips,kind="reg", height=4);
plt.show()

print("")
print("pairplot hue")
#sns.pairplot(data=tips,kind="reg", size=4, hue="sex"); # deprecated
sns.pairplot(data=tips,kind="reg", height=4, hue="sex");
plt.show()

print("")
print("pairplot smoker")
#sns.pairplot(data=tips,kind="reg", size=4, hue="smoker")
sns.pairplot(data=tips,kind="reg", height=4, hue="smoker");
plt.show()











```
