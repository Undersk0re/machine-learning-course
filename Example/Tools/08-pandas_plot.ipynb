# pie

```python
import pandas as pd

Data = {'Tasks': [300,500,700]}
df = pd.DataFrame(Data,columns=['Tasks'])
print(df.head())

# add column names
df = pd.DataFrame(Data,columns=['Tasks'], index = ['Tasks Pending','Tasks Ongoing','Tasks Completed'])
print(df.head())

df.plot.pie(y='Tasks',figsize=(5, 5),autopct='%1.1f%%', startangle=90);
#plt.show()
```

# Groupby

## PETS

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    'name':['john','mary','peter','jeff','bill','lisa','jose'],
    'age':[23,78,22,19,45,33,20],
    'gender':['M','F','M','M','M','F','M'],
    'state':['california','dc','california','dc','california','texas','texas'],
    'num_children':[2,0,0,3,2,1,4],
    'num_pets':[5,1,0,5,2,2,3]
})

preview_window=2
print("\n\n")
print("Preview dataset ------------------------")
print(df.head(preview_window))

print("")
print("Scatter")
df.plot(kind='scatter',x='num_children',y='num_pets',color='purple');
plt.show()

print("\n\n")
print("Lines children | pets ------------------------")
ax = plt.gca() # gca stands for 'get current axis
df.plot(kind='line',x='name',y='num_children',ax=ax)
df.plot(kind='line',x='name',y='num_pets', color='red', ax=ax);
plt.show()

print("\n\n")
print("Bar group by children ------------------------")
df.groupby('num_children').size().plot(kind='bar');
plt.show()

print("\n\n")
print("Bar group by pets ------------------------")
df.groupby('num_pets').size().plot(kind='bar');
plt.show()

print("\n\n")
print("Bar group by state ------------------------")
df.groupby('state').size().plot(kind='bar');
plt.show()

print("\n\n")
print("Bar group by state and gender ------------------------")
df.groupby(['state','gender']).size().plot(kind='bar');
plt.show()

print("\n\n")
print("Bar group by state and gender stacked ------------------------")
df.groupby(['state','gender']).size().unstack().plot(kind='bar',stacked=True);
plt.show()

print("\n\n")
print("Bar group by gender and state stacked with legend ------------------------")
df.groupby(['gender','state']).size().unstack().plot(kind='bar',stacked=True, figsize=(6, 6))
plt.legend(loc='upper left');
plt.show()
```
