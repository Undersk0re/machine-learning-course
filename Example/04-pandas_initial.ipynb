## Selection and Indexing

Let's learn how to retrieve information from a DataFrame.

### COLUMNS

We will begin be learning how to extract information based on the columns

```python
import pandas as pd

preview_window=2

url_tips="https://frenzy86.s3.eu-west-2.amazonaws.com/IFAO/tips_complete.csv"
df = pd.read_csv(url_tips)
print("\nViasualizza anteprima testa dataset ------------------------")
print(df.head(preview_window))

print("\nGrab a Single Column ------------------------")
print(df['total_bill'][:preview_window])

print("\nCheck type ------------------------")
print(type(df['total_bill'])) # pandas series == numpy array

print("\nCheck mean ------------------------")
print(df[['tip','total_bill']].mean()) 

print("\nReturn single named column ------------------------")
print(df.total_bill) 

print("\nGrab Multiple Columns ------------------------")
# Note how its a python list of column names! Thus the double brackets.
print(df[['total_bill','tip']][:preview_window])  # è una lista di selezione!!! ref SQL

print("\nCreate new column ------------------------")
df['new'] = 2
print(df.head(preview_window)) 

print("\nOperation by column ------------------------")
df['tip_percentage']=100* df['tip'] / df['total_bill']
df['price_per_person'] = df['total_bill'] / df['size']
print(df.head(preview_window)) 

print("\nAdjust Existing Columns ------------------------")
print("Because pandas is based on numpy, we get awesome capabilities with numpy's universal functions!")
df['price_per_person'] = np.round(df['price_per_person'],2)
print(df.head(preview_window)) 

print("\nRemove Columns ------------------------")
df = df.drop('tip_percentage',axis=1) #  #sovrascrivere!!!!!!!
print(df.head(preview_window)) 

print("\nIndex Basics ------------------------")
print("Before going over the same retrieval tasks for rows, let's build some basic understanding of the pandas DataFrame Index.")
print(df.index[:preview_window]) 
print("\nIndex Payment ID")
df = df.set_index('Payment ID') #sovrascrivere!!!!!!!
print(df.head(preview_window)) 
print("\nIndex tip")
print(df.set_index('tip').head(preview_window))  # does it make sense?
print("\nIndex reset")
df = df.reset_index()
print(df.index[:preview_window]) 


print("\nROWS ------------------------")
print("Let's now explore these same concepts but with Rows.")

print("\nGrab a Single Row")
print(df.iloc[0]) # prende la riga con indice indicato

print("\nName Based")
df = df.set_index('Payment ID')
print(df.loc['Sun2959']) # pesca dall'index


print("\nGrab Multiple Rows ------------------------")
print("\nprime righe")
print(df.iloc[0:4]) #prime 4 righe

print("\nprime per named index")
print(df.loc[['Sun2959','Sun5260']])
print("\nprime per step -2")
print(df.iloc[:,-2:].head(preview_window))
print("\ndrop per named index")
print(df.drop('Sun2959',axis=0).head(preview_window))


print("\nInsert a New Row ------------------------")
print("""Pretty rare to add a single row like this. 
Usually you use pd.concat() to add many rows at once. 
You could use the .append() method with a list of pd.Series() objects, 
but you won't see us do this with realistic real-world data.""")

print("\nprima riga separata")
one_row = df.iloc[0]
print(one_row)
print(type(one_row))
print("\nStampa ultime righe")
print(df.tail(preview_window))

print("\nStampa ultime righe con riga appesa")
# df.append(one_row).tail() # deprecato
df.loc[len(df)] = one_row
print(df.tail(preview_window))


print("\nExtract an array ------------------------")
array_row = df.iloc[0].values
print(array_row[:preview_window])

print("\nExtract a list ------------------------")
print(df['total_bill'].tolist()[:preview_window] )
```
