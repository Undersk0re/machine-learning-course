```python
import pandas as pd

data = {'Nomi': ['Alice', ' Bob', 'Gian Franco', 'David'],
        'Nomi_Cognomi': ['Alice Blues', ' Bob Cartney', 'GianFranco Porri', 'David Lee'],
        'Anni': [24,56,29,37]
        }

df = pd.DataFrame(data)
df
```

```python
#Aggiungi una nuova colonna (Lunghezza) al tuo DataFrame che contenga la lunghezza di ogni stringa in un'altra colonna
df['Lunghezza_nomi'] = df['Nomi'].astype(str).str.len()
df['Lunghezza_n_c'] = df['Nomi_Cognomi'].astype(str).str.len()
df
```

```python
# ripristina
df = df.drop('Lunghezza_nomi', axis=1) # asse 0 = x; asse 1 = y
df = df.drop('Lunghezza_n_c', axis=1)
df
```

```python
#Modifica tutte le stringhe in una colonna in maiuscolo(Nomi_Maiuscolo).
df['Nomi_Maiuscolo'] = df['Nomi'].astype(str).str.upper() # funzionale
df
```

```python
df['Lunghezza'] = df['Nomi'].astype(str).str.len()
df
```

```python
# Conta quante volte una particolare sottostringa appare in una colonna.(Conta_A)
#df['Conta_A'] = df['Nomi_Maiuscolo'].str.lower().str.count('a')
df['Conta_A'] = df['Nomi'].str.count('a')
df
```

```python
#Estrai i primi tre caratteri da ogni stringa in una colonna(Primi_3_Caratteri)
df['Primi_3_char'] = df['Nomi'].str[:3] # ->
df
```

```python
#Estrai gli ultimi 2 caratteri da ogni stringa in una colonna(Ultimi_2_Caratteri)
df['Ultimi_2_char'] = df['Nomi'].str[-2:]  # <-
df
```

```python
#Sostituisci una particolare sottostringa con un'altra in una colonna.(Nomi_Modificati)
df['Nomi_mod'] = df['Nomi'].str.replace('a' , 'X')
df
```

```python
#Aggiungi una colonna booleana (Contiene_C) che indica se una certa sottostringa Ã¨ presente in ogni stringa.
df['Contiene_C'] = df['Nomi'].str.find('c') # return position or negative
df.head()
```

```python
df['Contiene_C'] = df['Contiene_C'] > 0
df
```

```python
#Elimina gli spazi bianchi all'inizio e alla fine di ogni stringa in una colonna (Nomi_Senza_Spazi)
df['Nomi_Senza_Spazi'] = df['Nomi'].str.removeprefix(' ').str.removesuffix(' ')
df
```

```python
#Suddividi una colonna di stringhe in base a un delimitatore e crea nuove colonne.
print(pd.__version__)
#https://pandas.pydata.org/docs/user_guide/basics.html#tablewise-function-application
```

```python
df['Primo_Nome'] = df['Nomi_Senza_Spazi'].str.split(' ').str.get(0)
df
```

```python
df['Cognome'] = df['Nomi_Cognomi'].str.split(' ').str.get(-1)
df
```

```python
#Conta il numero di parole in ogni stringa di una colonna.(Conta_Parole)
df['Conta_Parole'] = df['Nomi_Senza_Spazi'].str.split(' ').str.len()
df
```

```python
#Combina piÃ¹ colonne di stringhe in una colonna formattata.
# https://pandas.pydata.org/docs/user_guide/merging.html
df['Nome_Completo'] = df['Nomi_Senza_Spazi'] + ' ' + df['Cognome']
df
```
